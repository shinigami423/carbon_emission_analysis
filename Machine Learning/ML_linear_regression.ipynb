```python
# Select the main "industr_type"
sectors = {"Utilities",
           "Waste Management and Remediation Services",
           "Oil and Gas Extraction",
           "Chemical Manufacturing ",
           "Pipeline Transportation",
          "Nonmetallic Mineral Product Manufacturing",
          "Primary Metal Manufacturing",
          "Petroleum and Coal Products Manufacturing"}

ML_optimized_df = ML_df.loc[ML_df["industry_type"].isin(sectors)]
ML_optimized_df
```

```python
# Select the columns we need for the linear regression
ML_optimized_df = ML_optimized_df.drop(columns = ["Unnamed: 0", "fac_id", "year", "State"], axis = 1)
ML_optimized_df
```

```python
# Transform the categorical features into numerical data
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
#ML_optimized_df['State'] = le.fit_transform(ML_optimized_df['State'])
ML_optimized_df['industry_type'] = le.fit_transform(ML_optimized_df['industry_type'])
ML_optimized_df
```

```python
from sklearn.preprocessing import OneHotEncoder

#categorical_transformer = OneHotEncoder(handle_unknown="ignore")
#ML_optimized_df['State'] = enc.fit(ML_optimized_df['State'])
#ML_optimized_df['sector'] = enc.fit_transform(ML_optimized_df['sector'])
#ML_optimized_df
```

```python
ML_optimized_df = ML_optimized_df.drop(columns = ["state_population"], axis = 1)
```

```python
# Define the features set
X = ML_optimized_df.copy()
X = X.drop("total_direct_emissions", axis=1)
X.head()
```

```python
# Define the target set.
y = ML_optimized_df["total_direct_emissions"].values
y[:5]
```

```python
# Split the Data Into Training and Testing Sets into an 80/20 split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, train_size=0.75)
```

```python
# Scale the Training and Testing Data
scaler = StandardScaler()

# Fit and scale the Standard Scaler with the training data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
#  Create the decision tree regressor instance
model = DecisionTreeRegressor(random_state=0)

# Fit Model
model = model.fit(X_train_scaled, y_train)
```

```python
print(model.score(X_train_scaled, y_train))
print(model.score(X_test_scaled, y_test))
```

```python
# Make predictions using the testing data
predictions = model.predict(X_test_scaled)
predictions
```

```python
# Test the balanced accuracy score of the model
from sklearn.metrics import r2_score
r2_score(y_test, predictions)
```
